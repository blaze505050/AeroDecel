Interactive exploration of planetary atmosphere models for Mars, Venus, and Titan.
From the AeroDecel v6.0 framework.


```python
import sys; sys.path.insert(0, "..")
import numpy as np
import matplotlib.pyplot as plt
from src.planetary_atm import get_planet_atmosphere
```

## Mars Atmosphere Profile
Multi-layer exponential model with CO2-dominated composition.


```python
mars = get_planet_atmosphere("mars")
h = np.linspace(0, 125_000, 500)

rho = [mars.density(hi) for hi in h]
T = [mars.temperature(hi) for hi in h]
P = [mars.pressure(hi) for hi in h]
a = [mars.speed_of_sound(hi) for hi in h]

fig, axes = plt.subplots(1, 4, figsize=(18, 5))
fig.suptitle("Mars Atmosphere Profile", fontweight="bold", fontsize=14)

axes[0].semilogy(rho, h/1e3, "r-", lw=2)
axes[0].set_xlabel("Density [kg/m3]")
axes[0].set_ylabel("Altitude [km]")
axes[0].grid(True, alpha=0.3)

axes[1].plot(T, h/1e3, "orange", lw=2)
axes[1].set_xlabel("Temperature [K]")
axes[1].grid(True, alpha=0.3)

axes[2].semilogy(P, h/1e3, "cyan", lw=2)
axes[2].set_xlabel("Pressure [Pa]")
axes[2].grid(True, alpha=0.3)

axes[3].plot(a, h/1e3, "lime", lw=2)
axes[3].set_xlabel("Speed of Sound [m/s]")
axes[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Comparative: All Three Planets


```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for name, color in [("mars", "red"), ("venus", "gold"), ("titan", "cyan")]:
    planet = get_planet_atmosphere(name)
    h = np.linspace(0, 125_000, 500)
    rho = [planet.density(hi) for hi in h]
    T = [planet.temperature(hi) for hi in h]
    axes[0].semilogy(rho, h/1e3, color=color, lw=2, label=name.title())
    axes[1].plot(T, h/1e3, color=color, lw=2, label=name.title())
    a = [planet.speed_of_sound(hi) for hi in h]
    axes[2].plot(a, h/1e3, color=color, lw=2, label=name.title())

for ax in axes:
    ax.set_ylabel("Altitude [km]")
    ax.legend(); ax.grid(True, alpha=0.3)
axes[0].set_xlabel("Density [kg/m3]")
axes[1].set_xlabel("Temperature [K]")
axes[2].set_xlabel("Speed of Sound [m/s]")
fig.suptitle("Comparative Planetary Atmospheres", fontweight="bold")
plt.tight_layout(); plt.show()
```
