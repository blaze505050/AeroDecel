Explore CO2 dissociation equilibrium and real-gas effects
during Mars atmospheric entry. Uses the Park (1993) 5-species model.


```python
import sys; sys.path.insert(0, "..")
import numpy as np
import matplotlib.pyplot as plt
from src.realgas_chemistry import RealGasChemistry
```

## CO2 Dissociation vs Temperature


```python
rg = RealGasChemistry()

T_range = np.linspace(1000, 8000, 200)
gamma_eff = [rg.effective_gamma(T) for T in T_range]
mass_fracs = [rg.equilibrium_composition(T) for T in T_range]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(T_range, gamma_eff, "cyan", lw=2.5)
ax1.axhline(1.4, color="red", ls="--", alpha=0.5, label="Ideal gas g=1.4")
ax1.axhline(1.28, color="orange", ls="--", alpha=0.5, label="CO2 cold g=1.28")
ax1.set_xlabel("Temperature [K]"); ax1.set_ylabel("Effective gamma")
ax1.set_title("Ratio of Specific Heats", fontweight="bold")
ax1.legend(); ax1.grid(True, alpha=0.3)

species = ["CO2", "CO", "O2", "O", "C"]
colors = ["red", "blue", "green", "orange", "purple"]
for sp, col in zip(species, colors):
    vals = [mf.get(sp, 0) for mf in mass_fracs]
    ax2.plot(T_range, vals, color=col, lw=2, label=sp)
ax2.set_xlabel("Temperature [K]"); ax2.set_ylabel("Mass Fraction")
ax2.set_title("Equilibrium Composition (Park 1993)", fontweight="bold")
ax2.legend(); ax2.grid(True, alpha=0.3)

plt.tight_layout(); plt.show()
```

## Stagnation-Point Heating Comparison
Compare Sutton-Graves (engineering) vs Fay-Riddell (exact) correlations.


```python
v_range = np.linspace(2000, 7000, 100)
rho = 5e-4  # typical density at peak heating altitude
R_n = 2.25  # nose radius [m]

q_sg = rg.sutton_graves(rho, v_range, R_n)
q_fr = rg.fay_riddell(rho, v_range, R_n)

plt.figure(figsize=(10, 5))
plt.plot(v_range/1e3, q_sg/1e4, "r-", lw=2, label="Sutton-Graves")
plt.plot(v_range/1e3, q_fr/1e4, "b--", lw=2, label="Fay-Riddell")
plt.xlabel("Velocity [km/s]"); plt.ylabel("Heat Flux [W/cm2]")
plt.title("Stagnation-Point Heating: Engineering Correlations", fontweight="bold")
plt.legend(); plt.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```
