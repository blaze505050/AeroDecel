Latin-Hypercube sampling for uncertainty propagation through the
EDL physics chain. Landing ellipse, sensitivity tornado, and

```python
import sys; sys.path.insert(0, "..")
import numpy as np
import matplotlib.pyplot as plt
from src.monte_carlo_edl import MonteCarloEDL
```

## Configure and Run Monte Carlo


```python
mc = MonteCarloEDL(
    planet_name="mars",
    n_samples=200,      # increase for production (1000+)
    verbose=True
)
results = mc.run()
print(f"Completed {results['n_samples']} samples")
print(f"Success rate: {results['success_rate']*100:.1f}%")
```

## Landing Distributions


```python
v_land = results["landing_velocities"]

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

axes[0].hist(v_land, bins=30, color="cyan", alpha=0.7, edgecolor="white")
axes[0].axvline(np.mean(v_land), color="red", ls="--", lw=2,
    label=f"Mean={np.mean(v_land):.0f} m/s")
axes[0].set_xlabel("Landing Velocity [m/s]")
axes[0].set_title("Landing Velocity Distribution", fontweight="bold")
axes[0].legend(); axes[0].grid(True, alpha=0.3)

if "peak_g_loads" in results:
    axes[1].hist(results["peak_g_loads"], bins=30, color="orange", alpha=0.7)
    axes[1].set_xlabel("Peak g-load")
    axes[1].set_title("Peak Deceleration", fontweight="bold")

if "flight_times" in results:
    axes[2].hist(results["flight_times"], bins=30, color="lime", alpha=0.7)
    axes[2].set_xlabel("Flight Time [s]")
    axes[2].set_title("Flight Time", fontweight="bold")

for ax in axes:
    ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```
